```python
import pandas as pd
import datetime
import numpy as np
```

```python
divvy_data = pd.read_csv("Downloads/202402-divvy-tripdata/202402-divvy-tripdata.csv")
```

```python
weather_data = pd.read_csv("Downloads/Chicago,United States 2024-02-01 to 2024-02-29.csv")
```

```python
divvy_data['started_at'] = pd.to_datetime(divvy_data['started_at'])
divvy_data['ended_at'] = pd.to_datetime(divvy_data['ended_at'])
weather_data['datetime'] = pd.to_datetime(weather_data['datetime'])

divvy_data.info()
weather_data.info()
```

```python
divvy_datetime = divvy_data[['started_at','ended_at']].drop_duplicates().reset_index(drop=True) 
divvy_datetime['start_second'] = divvy_datetime['started_at'].dt.second
divvy_datetime['start_minute'] = divvy_datetime['started_at'].dt.minute
divvy_datetime['start_hour'] = divvy_datetime['started_at'].dt.hour
divvy_datetime['start_day'] = divvy_datetime['started_at'].dt.day
divvy_datetime['start_month'] = divvy_datetime['started_at'].dt.month
divvy_datetime['start_year'] = divvy_datetime['started_at'].dt.year

divvy_datetime['end_second'] = divvy_datetime['ended_at'].dt.second
divvy_datetime['end_minute'] = divvy_datetime['ended_at'].dt.minute
divvy_datetime['end_hour'] = divvy_datetime['ended_at'].dt.hour
divvy_datetime['end_day'] = divvy_datetime['ended_at'].dt.day
divvy_datetime['end_month'] = divvy_datetime['ended_at'].dt.month
divvy_datetime['end_year'] = divvy_datetime['ended_at'].dt.year

divvy_datetime['ride_duration'] = divvy_datetime['ended_at'] - divvy_datetime['started_at']

divvy_datetime['divvy_datetime_id'] = divvy_datetime.index
divvy_datetime.head()
```

```python
divvy_datetime = divvy_datetime[['divvy_datetime_id', 'started_at','start_hour','start_minute', 'start_second','start_month', 'start_day', 'start_year', 'ended_at', 'end_hour', 'end_minute', 'end_second', 'end_month',  'end_day', 'end_year','ride_duration']]

divvy_datetime.head()
```

```python
weather_datetime = pd.DataFrame(weather_data['datetime'].drop_duplicates().reset_index(drop=True)) 
weather_datetime['recorded_hour'] = weather_data['datetime'].dt.hour
weather_datetime['recorded_day'] = weather_data['datetime'].dt.day
weather_datetime['recorded_month'] = weather_data['datetime'].dt.month
weather_datetime['recorded_year'] = weather_data['datetime'].dt.year

weather_datetime['weather_datetime_id'] = weather_datetime.index

weather_datetime.tail()
```

```python
divvy_data = divvy_data.merge(divvy_datetime)
weather_data = weather_data.merge(weather_datetime)
```
